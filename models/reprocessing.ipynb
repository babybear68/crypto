```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
btc = pd.read_csv('../raw data/new/btc_hourly.csv')
eth = pd.read_csv('../raw data/new/eth_hourly.csv')
ltc = pd.read_csv('../raw data/new/ltc_hourly.csv')
```

```python
def preprocessing(df):
    df['vol_per_count'] = df['volume_traded']/df['trades_count']
    for i in df.columns.tolist()[-4:]:
        df[i] = StandardScaler().fit_transform(df[i].values.reshape(-1, 1))
    df = df.iloc[:, -4:]
    return df
```

```python
btc = preprocessing(btc)
eth = preprocessing(eth)
ltc = preprocessing(ltc)
```
