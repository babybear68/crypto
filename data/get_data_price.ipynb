```python
timelist = []
for day in range(1, 31):
    for hour in range(24):
        string = '2019-04-'+str(day).zfill(2)+'T'+str(hour).zfill(2)
        timelist.append(string)
```

```python
BTC_dict = {}

for time in timelist:
    url = 'https://rest.coinapi.io/v1/exchangerate/BTC/USD?time='+time
    response = requests.get(url, headers=headers)
    BTC_dict[time] = response.json()['rate']
    print('successful: '+time)
    
BTC = pd.DataFrame.from_dict(BTC_dict, orient='index', columns=['price'])
```

```python
BTC.to_csv('BTC-USD_hour.csv')
```

```python
ETH_dict = {}

for time in timelist:
    try:
        url = 'https://rest.coinapi.io/v1/exchangerate/ETH/USD?time='+time
        response = requests.get(url, headers=headers)
        ETH_dict[time] = response.json()['rate']
        print('successful: '+time)
    except:
        print('failed: '+time)
    
ETH = pd.DataFrame.from_dict(ETH_dict, orient='index', columns=['price'])
```
