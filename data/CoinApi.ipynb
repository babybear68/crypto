```python
# import websockets
import datetime
import json
import requests
```

```python
key_file = 'CoinApiKey.json'
with open(key_file) as f:
    keys = json.load(f)
```

```python
headers = {'X-CoinAPI-Key' : keys['X-CoinAPI-Key']}
```

```python
url = 'https://rest.coinapi.io/v1/orderbooks/current'
response_curr = requests.get(url, headers=headers)
```

```python
response_curr.json()
```
